```python
from keras.preprocessing.image import ImageDataGenerator, img_to_array, load_img
from imutils import paths
import os
from keras_preprocessing import image
from shutil import copyfile
import tensorflow 

datagen = ImageDataGenerator(
    
#     adaptive_equalization=True, 
#         histogram_equalization=True,
        rotation_range=90,
#         brightness_range=[0.2,1],
        width_shift_range=0.1,
        height_shift_range=0.1,
        
        horizontal_flip=True,
        vertical_flip=True,
        fill_mode='nearest'
    
          )



```

```python

imagePaths = list(paths.list_images("./binaug_cervic_train/original/abnormal_old"))
print(imagePaths)



filepath='./binaug_cervic_train/original/abnormal/'


for (i, imagePath) in enumerate(imagePaths):
# extract the person name from the image path
#print("[INFO] processing image {}/{}".format(i + 1,len(imagePaths)))
#     dir1 = imagePath.split(os.path.sep)[-2]
    name1 = (imagePath.split(os.path.sep)[-1]).replace('.BMP','')
#     path=filepath
#     print(path)
    filename=filepath+'/'+name1+'.BMP'
    #print(filename)
    if not os.path.exists(filepath):
        os.makedirs(filepath)
    copyfile(imagePath, filename)
    img = load_img(imagePath)  
    x = img_to_array(img)  # creating a Numpy array with shape (3, 150, 150)
    x = x.reshape((1,) + x.shape)  # converting to a Numpy array with shape (1, 3, 150, 150)

    i = 0
    for batch in datagen.flow(x, save_to_dir=filepath, save_prefix=name1, save_format='.BMP'):
        i += 1
        if i >20:
            break 
```
